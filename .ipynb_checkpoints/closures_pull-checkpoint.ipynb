```python
import urllib.request, json 
with urllib.request.urlopen("https://newengland511.org/List/GetData/traffic?query=%7B%22columns%22%3A%5B%7B%22data%22%3Anull%2C%22name%22%3A%22%22%7D%2C%7B%22name%22%3A%22type%22%7D%2C%7B%22name%22%3A%22eventType%22%2C%22search%22%3A%7B%22value%22%3A%22Flood%22%7D%7D%2C%7B%22name%22%3A%22state%22%2C%22search%22%3A%7B%22value%22%3A%22Vermont%22%7D%7D%2C%7B%22name%22%3A%22county%22%7D%2C%7B%22name%22%3A%22roadwayName%22%7D%2C%7B%22name%22%3A%22direction%22%7D%2C%7B%22name%22%3A%22description%22%7D%2C%7B%22name%22%3A%22startTime%22%7D%2C%7B%22name%22%3A%22endTime%22%7D%2C%7B%22name%22%3A%22lastUpdated%22%7D%2C%7B%22data%22%3A11%2C%22name%22%3A%22%22%7D%5D%2C%22order%22%3A%5B%7B%22column%22%3A3%2C%22dir%22%3A%22asc%22%7D%5D%2C%22start%22%3A0%2C%22length%22%3A100%2C%22search%22%3A%7B%22value%22%3A%22%22%7D%7D&lang=en") as url:
    data = json.load(url)
    print(data)
```

```python
with urllib.request.urlopen("https://newengland511.org/map/mapIcons/Incidents") as url:
    print(url.read(500))
    incidents = json.load(url)
    print(incidents)
```

```python
with urllib.request.urlopen("https://newengland511.org/map/mapIcons/IncidentClosures") as url:
    incidentclosures = json.load(url)
    print(incidentclosures)
```

```python
incidents_df = pd.DataFrame.from_dict(incidents["item2"])
closures_df = pd.DataFrame.from_dict(incidentclosures["item2"])
combined_df = pd.concat([incidents_df, closures_df], axis=0)[["itemId", "location"]]
```

```python
fixed_columns = pd.DataFrame(combined_df.location.tolist(), index=combined_df.index, columns=["lat","lon"])
fixed_columns
combined_df["lat"] = fixed_columns["lat"]
combined_df["lon"] = fixed_columns["lon"]
```

```python
df = pd.DataFrame.from_dict(data["data"]).rename(columns={"DT_RowId" : "itemId"})
df
```

```python
merged = df.merge(combined_df, on="itemId")
merged
```

```python
merged.to_csv("data/closures.csv", index=False)
```
